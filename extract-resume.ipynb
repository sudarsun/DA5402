```python
import json
import pandas as pd
```

```python
# load the json file
with open('blah.json', 'r') as f:
    data = json.load(f)
```

```python
# create a spreadsheet with the following columns: filename, name, email, phone, location, linkedin, github
df = pd.DataFrame(columns=['filename', 'name', 'email', 'phone', 'location', 'linkedin', 'github'])
# loop through the data and extract the relevant information
for item in data:
    filename = item['file_name']
    name = item['name']
    email = ",".join(item['emails'])
    phone = ",".join(item['phones'])
    location = item['location']
    linkedin = item["links"]["linkedin"]
    github = item['links']['github']
    
    # add a row to the dataframe
    df.loc[len(df)] = [filename, name, email, phone, location, linkedin, github]
    #df = df.append({'filename': filename, 'name': name, 'email': email, 'phone': phone, 'location': location, 'linkedin': linkedin, 'github': github}, ignore_index=True)
# save the spreadsheetto a csv file
df.to_csv('resumes.csv', index=False)
```
